```python
import numpy as np
import matplotlib.pyplot as plt
```

## Regression Evaluation Methods

```python
## Our Data
n = 30

x = 10*np.random.randn(n)
y = -1.2*x + 40 + 4*np.random.randn(n)
y_pred = -1.2*x + 40

plt.scatter(x,y)
x_line = np.linspace(x.min(),x.max(),10)
plt.plot(x_line, -1.2*x_line + 40, color='red')
plt.show()
```

```python
abs_error = np.abs(y_pred-y)
sq_error = np.pow((y_pred-y),2)
sqlog_error = np.pow(np.log(y_pred+1) - np.log(y+1),2)

mae = np.mean(abs_error)
mse = np.mean(sq_error)
rmsle = np.sqrt(np.mean(sqlog_error))

fig, ax = plt.subplots(1,3, figsize=(12,5))

ax[0].scatter(x,abs_error, color='blue', label='Absolute Error')
ax[0].set_title(f"Absolute Error\nMAE = {mae:0.4f}\n")
ax[1].scatter(x,sq_error, color='red', label='Squared Error')
ax[1].set_title(f"Squared Error\nMSE = {mse:0.4f}\nRMSE = {np.sqrt(mse):0.4f}")
ax[2].scatter(x,sqlog_error, color='green', label='Sq Log Error')
ax[2].set_title(f"Squared Log Error\nRMSLE = {rmsle:0.4f}\n")

# for i in range(len(x)):
#     plt.plot([x[i], x[i]], [abs_error[i], sq_error[i]], color='grey', linestyle='dashed')
# plt.legend()

plt.show()
```

```python
x2 = np.append(x, -5)
y2 = np.append(y, 15)
y_pred2 = -1.2*x2 + 40

abs_error2 = np.abs(y_pred2-y2)
sq_error2 = np.pow((y_pred2-y2),2)
sqlog_error2 = np.pow(np.log(y_pred2+1) - np.log(y2+1),2)

mae2 = np.mean(abs_error2)
mse2 = np.mean(sq_error2)
rmsle2 = np.sqrt(np.mean(sqlog_error2))

fig, ax = plt.subplots(1,3, figsize=(12,5))

ax[0].scatter(x2,abs_error2, color='blue', label='Absolute Error')
ax[0].set_title(f"Absolute Error\nMAE = {mae2:0.4f}\n")
ax[1].scatter(x2,sq_error2, color='red', label='Squared Error')
ax[1].set_title(f"Squared Error\nMSE = {mse2:0.4f}\nRMSE = {np.sqrt(mse2):0.4f}")
ax[2].scatter(x2,sqlog_error2, color='green', label='Sq Log Error')
ax[2].set_title(f"Squared Log Error\nRMSLE = {rmsle2:0.4f}\n")

plt.show()
```
